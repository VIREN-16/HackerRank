```python
n = 5
#scores = list(map(int, input().split()))
scores = [2,3,6,6,5]

# Sort the list in descending order
sorted_scores = sorted(scores, reverse=True)

# Convert the sorted list into a set to remove duplicates
unique_sorted_scores = set(sorted_scores)

# Convert the set back into a list and access the second last element
runner_up_score = sorted(unique_sorted_scores)[-2]

print(runner_up_score)
```
