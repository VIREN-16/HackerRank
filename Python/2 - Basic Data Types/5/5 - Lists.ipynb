```python
if __name__ == '__main__':
    commands = [
        "insert 0 5",
        "insert 1 10",
        "insert 0 6",
        "print",
        "remove 6",
        "append 9",
        "append 1",
        "sort",
        "print",
        "pop",
        "reverse",
        "print"
    ]

    lst = []

    for cmd in commands:
        parts = cmd.split()

        if parts[0] == "insert":
            lst.insert(int(parts[1]), int(parts[2]))
        elif parts[0] == "print":
            print(lst)
        elif parts[0] == "remove":
            lst.remove(int(parts[1]))
        elif parts[0] == "append":
            lst.append(int(parts[1]))
        elif parts[0] == "sort":
            lst.sort()
        elif parts[0] == "pop":
            lst.pop()
        elif parts[0] == "reverse":
            lst.reverse()
```
