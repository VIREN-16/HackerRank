```python
# Step 1: Read input
n = int(input())
students = [[input(), float(input())] for _ in range(n)]

# Step 2: Find the second lowest grade
second_lowest_grade = sorted(set(grade for name, grade in students))[1]

# Step 3: Print names of students with the second lowest grade
for name, grade in sorted(students):
    if grade == second_lowest_grade:
        print(name)
```

## OR

```python
if __name__ == '__main__':
    Students = []
    for _ in range(int(input())):
        name = input()
        score = float(input())
        Students.append([name,score])

sort_score = sorted(set(score for _, score in Students))      
second_lowest_grade = sort_score[1]  
second_lowest_student = sorted([name for name, score in Students if score == second_lowest_grade])

for Students in second_lowest_student:
    print(Students)
```
