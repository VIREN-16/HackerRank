```python
# Given example
string = "ABCDCDC"
substring = "CDC"

count = 0

# Traverse the string
for i in range(len(string) - len(substring) + 1):
    if string[i:i+len(substring)] == substring:
        count += 1

print(count)
```

## OR

```python
def count_substring(string, sub_string):
    count=0
    for i in range(len(string)):
        if string[i : i+len(sub_string)]==sub_string:
            count+=1
    return count
```
