```python
def swap_case(s):
    result= ""
    for i in s:
        if i.isupper():
            result += i.lower()
        else:
            result += i.upper()
    return result

s = 'HackerRank.com presents "Pythonist 2".'

swap_case(s)
```

## OR

```python
def swap_case(s):
    return s.swapcase()

# given input
s = 'HackerRank.com presents "Pythonist 2".'

# output
print(swap_case(s))
```
